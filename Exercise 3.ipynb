```
#Program to take in length of run times

print("This program calculates your average run time")

runtimes = []

runtime = input("Please enter your runtime in minutes ")

while runtime:
  runtimes.append(runtime)
  runtime = input("Please enter a run time in minutes ")

runtimes

runfloats = []
#Convert list of characters to list of floats
[runfloats.append(float(everyvalue)) for everyvalue in runtimes]


print(f"Your total time is {sum(runfloats)} and your average is {sum(runfloats)/len(runtimes)}")
```

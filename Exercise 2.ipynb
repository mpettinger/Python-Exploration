```
#
print("This program will take a variable number of arguments and return the sum of those numbers")
userinput = input("Please enter your input ")
userlist = userinput.split()

usernumbers=[]

for everych in userlist:
  usernumbers.append(int(everych))

thesum=0

for everynumber in usernumbers:
  thesum = everynumber + thesum

print(f"{thesum}")
```

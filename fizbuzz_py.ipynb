```
#Create the range of numbers to evaluate
begining_number = int(input("What is the starting number to evaluate?"))
ending_number = int(input("What is the ending number to evaluate?"))

#Create a list of numbers and words to evaluate for each number in the range
size_of_list = int(input("How many numbers will be evaluated?"))
eval_dict={}

for number in range(1,size_of_list+1):
  key = int(input("Enter divisor: "))
  value = input("Enter string for divisor: ")
  eval_dict[key] = value


def number_evaluator(e_number):
  return_string = ''
  for k, v in eval_dict.items():
    if e_number % k == 0:
      return_string = return_string + v
  if return_string == '':
    return_string = f"{e_number}"
  return return_string

for number in range(begining_number, ending_number+1):
  print(number_evaluator(number))
#print(eval_dict)

#Evaluate each number and associated word.

#Print each word or number.
```
